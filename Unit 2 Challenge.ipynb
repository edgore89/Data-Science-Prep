```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Here I compared the age and charges data using a scatter. 
#We can see that the price goes up as the age increases for each age group.
df=pd.read_csv('insurance.csv')

df.plot(kind='scatter',x='age',y='charges', color ='green')
plt.xlim([18,65])
plt.ylim([1100,64000])
plt.show()
```

```python
#Here I used a bar chart to determine if being a smoker or nonsmoker affects the charges of medical insurance.
#As we can see, people that smoke end up being charged more for medical insurance.
#we can also see that there are many more smokers than nonsmokers.
df.groupby('smoker').charges.mean().plot(kind ='bar')
plt.show()
```

```python
#Here I used a box plot to determine if being a smoker or nonsmoker affects the charges of medical insurance.
#As we can see, people that smoke end up being charged more for medical insurance.
#By using a box plot, we can see a bigger difference between the max and min charges values between smokers and non smokers.
sns.boxplot(x='smoker', y='charges', data=df)
```

```python
#Here I used a swarm plot to seprate the different regions data into sub groups to compare the insurance charges.
#By seprating the four regions, we can see how much each region pays for insurance and can compare the extrema.
sns.swarmplot(x='region', y='charges', data=df)
```
